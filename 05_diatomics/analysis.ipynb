```python
import qctoolkit as qtk
import numpy as np
from glob import glob
import pandas as pd
```

```python
qtk.setting.quiet=True
outs = []
for o in sorted(glob('/home/samio/data/Basel/Xe2Scan/Xe2Scan_def2q/*/*/*.log')):
    out = qtk.QMOut(o, program='gaussian')
    outs.append(out)
qtk.setting.quiet=False
```

```python
qtk.setting.quiet=True
atoms = []
for o in sorted(glob('/home/samio/data/Basel/Xe2Scan/freeAtom_ref/*/*.log')):
    out = qtk.QMOut(o, program='gaussian')
    atoms.append(out)
qtk.setting.quiet=False

akey = [a.name[3:-4] for a in atoms]
```

```python
atom_ref = {}

Z_range = range(1, 55)
for Z in Z_range:
    A = qtk.Z2n(Z)
    txt = filter(lambda x: A + '_' in x, akey)
    ind0, ind1 = akey.index(txt[0]), akey.index(txt[1])
    Ea1, Ea2 = atoms[ind0].Et, atoms[ind1].Et
    atom_ref[A] = min(Ea1, Ea2)
```

```python
columns = ['name', 'multiplicity', 'A1', 'A2', 'd', 'E_total', 'E_atomization']
data = []
for i, out in enumerate(outs):
    if out.molecule.N == 2:
        Z1, Z2 = out.molecule.Z[0], out.molecule.Z[1]
        A1, A2 = qtk.Z2n(Z1), qtk.Z2n(Z2)
        m = int(out.name.split("_m")[-1].split(".")[0])
        Ea = out.Et - atom_ref[A1] - atom_ref[A2]
        row = [out.name, m, A1, A2, out.molecule.R[0,0], out.Et, Ea]
        data.append(row)
```

```python
df = pd.DataFrame(data, columns=columns)
```

```python
df.to_csv('diatomic_ccsdt_devq.csv')
```
